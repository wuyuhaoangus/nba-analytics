```python
#function to get the team_id in order for us to retrieve the team stats
def get_team_id(team_name_abbrev):
    nba_teams=teams.get_teams()
    return [team for team in nba_teams
         if team['abbreviation'] == team_name_abbrev][0]['id']
```
